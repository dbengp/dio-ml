```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
#deve-se importar uma imagem a raiz desse diretório, que será carregada pela funcção cv.imread("imagem_importada.jpg")

bgr_im = cv.imread("gato-de-casa.jpg")

rgb_im = cv.cvtColor(bgr_im, cv.COLOR_BGR2RGB)
gray_im = cv.cvtColor(bgr_im, cv.COLOR_BGR2GRAY)

# Para obter uma imagem binária (preto e branco), deve-se aplicar um processo de limiarização (thresholding).
ret, thresh = cv.threshold(gray_im, 127, 255, cv.THRESH_BINARY)
img_array = np.array(thresh)

# cria os subplots para exibir as imagens lado a lado
plt.subplot(1, 3, 1), plt.imshow(rgb_im)
plt.title('Cores'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 3, 2), plt.imshow(gray_im, cmap='gray')
plt.title('Tons de Cinza'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 3, 3), plt.imshow(img_array, cmap='gray')
plt.title('Preto e Branco'), plt.xticks([]), plt.yticks([])
plt.show()


```
